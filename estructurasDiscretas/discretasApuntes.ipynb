```python
def fibo_iter(n):# iterativa ya sea for, while etc
    a, b = 1, 1 # tuple python
    for k in range(n):
        a, b = b, a+b # primer valor se lo asigna a a
    return a
print(fibo_iter(4))
```

### Hanoi

Consta en mover discos de una posición A a una posición C, usando los siguientes pasos:

1) Mover uno ala vez.
   
2) Se debe mover el disco que este en la cima (el que esta por encima de los demas).
   
3) Los discos grandes tienen colocados encima discos mas pequeños.

<img src="https://i.imgur.com/McxBfPp.jpg" width=""/>

Ejercicio codigo Hanoi

```python
def Hanoi(n, A = 'A', B = 'B', C= 'C'):
    if n == 1:
        print(f"Mueva de{A} a {C}")
        return
    Hanoi(n-1, A, C, B)
    print(f"Mueva de{A} a {C}")
    Hanoi(n-1, B, A, C)
print(Hanoi(2))
```

Consta en tomar un problema y dividirlo en problemas iguales.

```python
#DyC
def sumDyC(a:list[int]) -> int:
    if len(a) == 1:  return a[0] # returna 0 porque si es un elemento no puede subdividirse en dos, es el caso base
    divide = len(a)// 2    # tomamos el tamnio de la lista
    return sumDyC(a[0: divide]) + sumDyC(a[divide:]) # subdividimos el problema en dos mitades, hasta que ya no se pueda subdividir
                                                         #y unimos el resultado de ambas mitades consecutivamente 
a = [2,3,4]
print(sumDyC(a))
```

Ejemplo: merge

```python
def merge(a, b):
    n, m = len(a), len(b)
    result = [0] * (n+m)
    i, j, k = 0, 0, 0;
    
    while i < n and j < m:
        if a[i]< b[j]:
            result[k] = a[i]
            i += 1
        else:
            result[k] = b[j]
            j += 1
        k += 1
    while i < n:
        result[k] = a[i]
        i += 1
        k += 1
    while j < m:
        result[k] = b[j]
        j += 1
        k += 1
    return result

a = [10, 15, 16]
b = [1, 2, 14, 25, 26, 30]
print(merge(a, b))
```

```python
#practicas
def quickSort(a:list[int],izquierda, derecha) -> int:
    if izquierda < derecha:
        mitad = particion(a, izquierda, derecha)
        quickSort(a, izquierda, mitad)
        quickSort(a, mitad + 1, derecha)


def particion(a:list[int],izq,derec) -> int:
    pivote = a[izq]
    while True:
        while a[izq] < pivote:
            izq += 1
        while a[derec] > pivote:
            derec -= 1
        if izq >= derec: 
            return derec
        else:
            a[izq], a[derec] = a[derec], a[izq]
        izq  += 1
        derec -= 1
        
a = [5, 1, 2, 1, 1, 3, 5, 1, 5, 1, 99, 231, 234, 12, 121 ]
print("Antes de ordenarlo: ")
print(a)
quickSort(a, 0, len(a) - 1)
print("Después de ordenarlo: ")
print(a)
```

## Formalización asintótica: O(), Ω(), Θ()

O() f ~ O(g) -> ∃ n0 >= 0, ∃ c > 0 : f(n) <= cg(n) ∀ n >= n0

Una función g multiplicada por una constante en un punto (n0) es >= a f.

Se concluye que f crece a lo más como g basado por una constante

Ω() f ~ Ω(g) -> g ~ O(f)

Una función g multiplicada por una constante g sea < f basado en un punto n0.

Se concluye que f crece al menos como g basado por una constante, f >= cg

Θ() f ~ Θ(g)  -> f ~ Ω(g) ^ f ~ O(g)

Una función que al probar los dos casos, tiene una cota O(g) y Ω(g) al mismo tiempo, dandose dos funciones g, con constantes diferentes y diferentes puntos n0.

#### Teoremas importantes

1) c ~ O(1) ∀ c > 0
   
2) f1 + f2 ~ O( max(f1, f2) )

3) f1 * f2 ~ O( O(f1), O(f2) )

4) si es un polinomio, f ~ O(nd), se usa al monomio más grande

```python
import os
clear = lambda: os.system('cls')

class Relation:
    def __init__(self, A:set, B:set) : # Métodos especiales se denominan Dunder (Constructor).
        self.A = set(A) # Atributo A, se iguala al set que recibe por parámetro.
        self.B = set(B) # Atributo B, se iguala al set que recibe por parámetro.
        self.pairs = set() # Atributo pairs, set de todos los pares de la relación.
    
    def relate(self, x, y) : # Añade un par (x,y) que se encuentra relacionado al set pairs.
        print(f"{x, y} have been related!")
        self.pairs.add((x, y))
    
    def are_related(self, x, y) : # Verifica si un par (x,y) existe en el conjunto de elementos relacionados.
        return (x, y) in self.pairs
    
    def neighbors(self, x) : # Devuelve aquellos pares en los que x está presente.
        return {p[1] for p in self.pairs if p[0] == x}

class BinarySelfRelation(Relation):
    def __init__(self, A:set) : # Recibe un set cualquiera para relacionarlo consigo mismo.
        super().__init__(A, A) # Crea un objeto temporal según el padre, que permite acceder a los métodos del padre.
        

def is_transitive(R:BinarySelfRelation):
    for x in R.A : # Para cada 'x' en en la relación.
        for y in R.neighbors(x) : # Devuelve cuáles 'y' se relacionan con 'x'.
            for z in R.neighbors(y) : # Devuelve cuáles 'z' se relacionan con 'y'.
                if not R.are_related(x, z) : # Por definición debería existir una relación entre 'x' y 'z' según 'y'.
                    return (x, y, z) # Devuelve un contraejemplo.
    return True

def is_reflexive(R:BinarySelfRelation):
    for x in R.A :
        if not R.are_related(x, x) :# si cada x en A no se relaciona con si mismo, muestra el x
            return(x)
    return True

def is_symmetric(R:BinarySelfRelation):
    for x in R.A :
        for y in R.neighbors(x) : 
            if R.are_related(x, y) and not R.are_related(y, x) :# si un x, y que pertenecen a A, incumple que xRy ^ yRx, muestra x,y
                return (x, y)
    return True

def is_equivalent(R:BinarySelfRelation):
    if is_transitive(R) == True and is_reflexive(R) == True and is_symmetric(R) == True :#si cumple ser reflexiva,simetrica,transitiva da true
        return True
    return False

test_set:set = ('a', 'b', 'c', 'd')
test_self_relation = BinarySelfRelation(test_set)
test_self_relation.relate('a', 'a')
test_self_relation.relate('a', 'b')
test_self_relation.relate('a', 'd')
test_self_relation.relate('b', 'a')
test_self_relation.relate('b', 'b')
test_self_relation.relate('c', 'c')
test_self_relation.relate('d', 'a')
test_self_relation.relate('d', 'd')
clear()

print(f"{is_transitive(test_self_relation) = }")

print(f"{is_reflexive(test_self_relation) = }")

print(f"{is_symmetric(test_self_relation) = }")
```

```python
# ++++++++++++++++++ Cerraduras ++++++++++++++++++ 

def translate_to_matrix(R:BinarySelfRelation) -> list : # Construye una matriz según una BinarySelfRelation.
    new_matrix:list = []
    elements:list = list(R.A) # Los elementos siempre serán distintos.
    print(f"{elements = }") # MOSTRAR EL ORDEN EN EL CUAL QUEDA.
    for i in range (len(elements)) :
        new_matrix.append([0]*len(elements))   
    for i in range (len(elements)) :
        for j in range(len(elements)) :
            if R.are_related(elements[i], elements[j]) :
                new_matrix[i][j] = 1
    return new_matrix
  
def warshall_reflexive(R:BinarySelfRelation) :
    warshall_matrix:list = translate_to_matrix(R)
    for i in range(len(R.A)) :
        warshall_matrix[i][i] = 1
    return warshall_matrix
 
def warshall_symmetric(R:BinarySelfRelation) :
    warshall_matrix:list = translate_to_matrix(R)
    for i in range(len(R.A)) :
        for j in range(len(R.A)) :
            if warshall_matrix[i][j] != warshall_matrix[j][i] :
                warshall_matrix[i][j] = 1
                warshall_matrix[j][i] = 1        
    return warshall_matrix
 
def warshall_transitive(R:BinarySelfRelation) :
    warshall_matrix:list = translate_to_matrix(R)
    vertices = len(R.A)
    for k in range(vertices) :
        for i in range(vertices) :
            for j in range(vertices) :
                warshall_matrix[i][j] = warshall_matrix[i][j] or (warshall_matrix[i][k] and warshall_matrix[k][j])
    return warshall_matrix          
    
```

### Árbol balanceado (AVL )

Árbol que esta equilibrado, no esta más cargado de un lado que de otro.

Si la altura de la izquierda - la altura de la derecha es mayor a 1 , el árbol esta desiquilibrado.

La altura tiene que ser 0 o 1 en valor absoluto.

Pasos:

1) Inserción: inserta normal menores a la izquierda, mayores a la derecha.
   
2) Si se desequilibra aplica rotaciónes: simples si son recto, dobles si es sixac

Variables a utilizar:

w = último nodo inserta (hoja)

z = desde w hacia arriba el primero desequilibrado.

y = hijo de z.

x = hijo de y.

Tipos de rotaciónes simples:

SL = simple a la izquierda

SR = simple a la derecha.

Tipos de rotaciónes dobles:

Doble a la izquierda: SL y SR

Doble a la derecha: SR Y SL

Movimientos:

SI la rotación es simple y el árbol esta hacia la derecha = SL y/z

SI la rotación es simple y el árbol esta hacia la izquierda = SR y/z

SI la rotación es doble y el árbol esta en sixac hacia izquierda = 1° SL x/y, 2° SR x/z

SI la rotación es doble y el árbol esta en sixac hacia derecha = 1° SR x/y, 2° SL x/z
